```python
dataset_new= Data.dropna()
print(dataset_new.shape)
dataset_new
```

```python
#filling null values with a single value(matn ya adad)
new_dataset = Data.fillna(0)
new_dataset
```

```python
#filling null values with previous values(moshkelesh radife avale)

new= Data.fillna(method= 'pad')
new
```

```python
#filling null values with next values(moshkelesh radife akhare)

new= Data.fillna(method= 'bfill')
new
```

```python
#filling null value with the median(baraye har sotun hesab mikone)
new= Data.copy()
missing_columns = new.isnull().any()
for column in missing_columns[missing_columns].index:
  new[column].fillna(new[column].median(), inplace=True)
new
```

```python
new= Data.copy()
missing_columns = new.isnull().any()
for column in missing_columns[missing_columns].index:
  new[column].fillna(new[column].mode()[0], inplace=True) #0 ro mizrim ke bege avalin por tekrari ke peida mikoni ro bezar
new
```

```python
new= Data.copy()
missing_columns = new.isnull().any()
for column in missing_columns[missing_columns].index:
  new[column].fillna(new[column].mean(), inplace=True)
new
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```
